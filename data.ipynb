```python
X = df_temp_onehot.drop('SalePrice', axis=1).values
y = df_temp_onehot['SalePrice'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print('Shape')
print('X_train: ', X_train.shape)
print('X_test: ', X_test.shape)
print('y_train: ', y_train.shape)
print('y_test: ', y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logisticreg = LogisticRegression()
logisticreg.fit(X_train,y_train)
y_pred_logreg = logisticreg.predict(X_test)
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred_logreg)*100)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor()
from sklearn import metrics
dtree.fit(X_train,y_train)
y_pred_dtree = dtree.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred_dtree)*100)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
y_pred_svc = svc.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred_svc)*100)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_pred_gnb = gnb.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred_gnb)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtreeclas = DecisionTreeClassifier(max_depth=6, random_state=123,criterion='entropy')
dtreeclas.fit(X_train,y_train)
y_pred_dtc = dtree.predict(X_test)
print(metrics.accuracy_score(y_test, y_pred_dtc)*100)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(125, activation='relu'))
model.add(Dense(63, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(2, activation='relu'))

model.add(Dense(1))
# optimizer = 'adam' - regression model
model.compile(optimizer='adam', loss='mse')
```

```python
# validation_data is used to check the error, but does not change the weights of the model
model.fit(x = X_train, y = y_train,
          validation_data=(X_test, y_test),
          batch_size=128, epochs=1000)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
# the graph denotes that the loss vs val_loss is similar and reducing as the epochs move higher. The model can be trained further without overfitting.
```

```python
predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
print(confusion_matrix(y_test, predictions))
```
